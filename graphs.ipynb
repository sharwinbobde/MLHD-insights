```python
import pandas as pd
years = ["2008"]
```

```python
from glob import glob
import matplotlib.pyplot as plt

for year in years:
    for f in glob("out_data/yr_" + year + ".csv/part-*.csv"):
        df = pd.read_csv(f)
        #  drop 0 listens
        df = df.drop(index=0) 

        fig = plt.figure(figsize=(10,6))
        fig.suptitle("User-Recording listening frequency (year "+year+")", fontsize=16)
        plt.scatter(df["yr_" + year + "_listens"], df["count"], label = "count")
        plt.xscale('log')
        plt.yscale('log')
        plt.xlabel("Number of listens (log-scale)", fontsize=12)
        plt.ylabel("Frequency of occurance (log-scale)", fontsize=12)
        plt.legend()
```

```python
from glob import glob
import matplotlib.pyplot as plt

for year in years:
    for f in glob("out_data/yr_" + year + ".csv/part-*.csv"):
        df = pd.read_csv(f)
        #  drop 0 listens
        df = df.drop(index=0, axis=1) 

        fig = plt.figure(figsize=(10,6))
        fig.suptitle("User-Recording listening continuation into next year (year "+year+" to "+str(int(year)+1)+")", fontsize=16)
        plt.xscale('log')
        plt.yscale('symlog')
        plt.scatter(df["yr_" + year + "_listens"], df["next_yr_mean"], label = "mean")
        plt.scatter(df["yr_" + year + "_listens"], df["next_yr_std"], label = "standard deviation")
        plt.xlabel("Number of listens in "+year+" (log-scale)", fontsize=12)
        plt.ylabel("Number of listens in "+str(int(year)+1)+" (symmetric-log-scale)", fontsize=12)
        plt.legend()
```
