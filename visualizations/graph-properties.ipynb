```python
import json
import matplotlib.pyplot as plt
import seaborn as sns

with open('../graph_properties.json', 'rb') as f:
    data = json.load(f)
data.keys()
```

```python
sns.set_context('notebook', rc={'figure.figsize': (100,10)})
```

```python
plt.figure(figsize=(10,8))
sns.barplot(['users', 'artists', 'recordings'],
 [data['users'], data['artists'], data['recs']])
```

```python
plt.figure(figsize=(10,8))
sns.barplot(['users_to_recs', 'users_to_artists', 'artists_to_recs'],
 [data['users_to_recs'], data['users_to_artists'], data['artists_to_recs']])
```

```python
fig = plt.figure(figsize=(20,8))
fig.suptitle("users_to_recs")
for set_size in reversed(range(10, 144, 20)):
    sum_set = 'users_to_recs_upto_part_'+str(set_size)+'_sums'
    keys = sorted(data[sum_set].keys())
    heights = []
    for key in keys:
        heights.append(data[sum_set][key])
    plt.bar(keys, heights, label="upto "+str(set_size)+" parts")
plt.legend()
```
